## Lists

Lists are ordered collections of items. They are mutable, which means you can change them after they have been created. You can add, remove, and modify items in a list. They can contain duplicate items. Lists are defined by using square brackets `[]` and separating items with commas. A list can contain items of different types, including other lists. That means you can build complex data structures like matrices and trees using lists. Lists are indexed. The first item has an index of 0, the second item has an index of 1, and so on. You can also use negative indexing to access items from the end of the list. The last item has an index of -1, the second to last item has an index of -2, and so on.

Ordered in the context of a list means the order is defined by the index of the items. Once added to the list, an item will not change its position/index unless you explicitly change it.

```python
# Example of a list
my_list = [1, 2, 3, 'four', [5, 6]]
print(f"{my_list=}")

# Length of the list:
print(f"{len(my_list)=}")

# You can create lists or cast other data structures to lists with the `list()` function:
my_tuple = (1, 2, 3)
my_list_from_tuple = list(my_tuple)
print(f"{my_list_from_tuple=}")

# Access items in the list by index:
print(f"{my_list[0]=}")  # First item
print(f"{my_list[1]=}")  # Second item
print(f"{my_list[-1]=}")  # Last item

# Slicing lists:
# The fist number is the start index (inclusive) and the second number is the end index (exclusive).
# Slicing always creates a new list. It does not modify the original list.
print(f"{my_list[1:4]=}")  # Items from index 1 to 3
print(f"{my_list[:3]=}")  # First three items
print(f"{my_list[3:]=}")  # Items from index 3 to the end
print(f"{my_list[:]=}")  # Copy of the whole list
print(f"{my_list[::2]=}") # Step in slicing
print(f"{my_list[::-1]=}") # Reverse the list
print(f"{my_list[::-2]=}") # Reverse the list with step

# Linear search in a list:
if 'four' in my_list:
    print("Found 'four' in the list!")

# Change items in the list:
my_list[0] = 'one'
print(f"{my_list=}")

# Insert items in the list:
my_list.insert(1, 'two')
print(f"{my_list=}")

# Append items to the end of the list:
my_list.append('seven')
print(f"{my_list=}")

# Join two lists:
another_list = ['eight', 'nine']
joined_list = my_list + another_list
print(f"{joined_list=}")

# Remove items from the list by value:
# Only the first occurrence of the item will be removed.
my_list.remove('two')
print(f"{my_list=}")

# Remove items from the list by index:
# The `pop()` method removes and returns the item at the specified index. If no index is specified, it removes and returns the last item.
removed_item = my_list.pop(0)  # Remove the first item
print(f"{removed_item=}")
print(f"{my_list=}")

# Clear the list:
my_list.clear()
print(f"{my_list=}")

# List comprehensions is a concise way to create lists:
# Create a list of even squares of numbers from 0 to 9:
squares = [x**2 for x in range(10) if x % 2 == 0]  # Only include even numbers
print(f"{squares=}")

# Sorting a list ascending:
numbers = [5, 2, 9, 1, 5, 6]
numbers.sort()  # Sort the list in place
print(f"{numbers=}")

# Sorting a list descending:
numbers.sort(reverse=True)  # Sort the list in place in descending order
print(f"{numbers=}")

# Unpacking a list:
a, b, c, d, e, f = numbers
print(f"{a=}, {b=}, {c=}, {d=}, {e=}, {f=}")
```

## Tuples

A tuple is ordered by **unchangeable**. In Python tuples are written with round brackets `()`. Tuples are indexed. The first item has an index of 0, the second item has an index of 1, and so on. You can also use negative indexing to access items from the end of the tuple. The last item has an index of -1, the second to last item has an index of -2, and so on. Duplicate items are allowed in tuples. Unchangeable in the context of a tuple means that once a tuple is created, you cannot change its items. You cannot add, remove, or modify items in a tuple.

You can do most of the things with tuples that you can do with lists. You can access items, slice them, iterate over them, and so on. The main difference is that you cannot change a tuple after it has been created.

When to use a tuple instead of a list? You can use a tuple when you want to create a collection of items that should not be changed. For example, you can use a tuple to represent the coordinates of a point in 2D space, or the RGB values of a color. You can also use a tuple as a key in a dictionary, because tuples are hashable, while lists are not.

```python
# Example of a tuple
my_tuple = (1, 2, 3, 'four', [5, 6])
print(f"{my_tuple=}")

# Another notation:
my_tuple = 1, 2, 3, 'four', [5, 6]
print(f"{my_tuple=}")

# Length of the tuple:
print(f"{len(my_tuple)=}")

# You can create tuples or cast other data structures to tuples with the `tuple()` function:
my_list = [1, 2, 3]
my_tuple_from_list = tuple(my_list)
print(f"{my_tuple_from_list=}")

# Access items in the tuple by index:
print(f"{my_tuple[0]=}")  # First item
print(f"{my_tuple[1]=}")  # Second item
print(f"{my_tuple[-1]=}")  # Last item

# Slicing tuples:
# The fist number is the start index (inclusive) and the second number is the end index (exclusive).
# Slicing always creates a new tuple. It does not modify the original tuple.
print(f"{my_tuple[1:4]=}")  # Items from index 1 to 3
print(f"{my_tuple[:3]=}")  # First three items
print(f"{my_tuple[3:]=}")  # Items from index 3 to the end
print(f"{my_tuple[:]=}")  # Copy of the whole tuple
print(f"{my_tuple[::2]=}") # Step in slicing
print(f"{my_tuple[::-1]=}") # Reverse the tuple
print(f"{my_tuple[::-2]=}") # Reverse the tuple with step

# Linear search in a tuple:
if 'four' in my_tuple:
    print("Found 'four' in the tuple!")

# Join two tuples:
another_tuple = ('eight', 'nine')
joined_tuple = my_tuple + another_tuple
print(f"{joined_tuple=}")

# Unpacking a tuple:
a, b, c, d, e = my_tuple
print(f"{a=}, {b=}, {c=}, {d=}, {e=}")
```

## Sets

A set is an unordered collection of unique items. They are mutable, which means you can change them after they have been created. You can add and remove items from a set, but you cannot modify the items themselves. Sets are defined by using curly braces `{}` and separating items with commas. A set can contain items of different types, but they must be immutable (like numbers, strings, and tuples). Sets are not indexed, which means you cannot access items by their position. You can check if an item is in a set using the `in` keyword.

```python
# Example of a set:
my_set = {1, 2, 3, 'four', (5, 6)}
print(f"{my_set=}")

# Length of the set:
print(f"{len(my_set)=}")

# You can create sets or cast other data structures to sets with the `set()` function:
my_tuple = (1, 2, 3)
my_set_from_tuple = set(my_tuple)
print(f"{my_set_from_tuple=}")

# Set can be used to distinct items in a list:
my_list = [1, 2, 2, 3, 4, 4, 5]
my_set_from_list = set(my_list)
print(f"{my_set_from_list=}")

# Access items in the set by membership:
print(f"{1 in my_set=}")  # Check if 1 is in the set
print(f"{4 in my_set=}")  # Check if 4 is in the set

# Insert items in the set:
my_set.add('two')
print(f"{my_set=}")

# Remove items from the set by value:
# The `remove()` method removes the specified item from the set. If the item is not found, it raises a KeyError.
my_set.remove('two')
print(f"{my_set=}")

# Remove items from the set by value without raising an error if the item is not found:
my_set.discard('two')
print(f"{my_set=}")

# Join two sets:
# Update modifies the original set by adding items from another set. It does not create a new set.
another_set = {'eight', 'nine'}
my_set.update(another_set)
print(f"{my_set=}")

# With union() you can create a new set that contains all items from both sets:
union_set = my_set.union(another_set)
print(f"{union_set=}")

# Keeping only the duplicate items in both sets:
set_1 = {1, 2, 3, 4}
set_2 = {3, 4, 5, 6}
intersection_set = set_1.intersection(set_2)
print(f"{intersection_set=}")

# Keeping only the items from set_1 that are not in set_2:
difference_set = set_1.difference(set_2)
print(f"{difference_set=}")

# Keeping only the items that are not in both sets:
symmetric_difference_set = set_1.symmetric_difference(set_2)
print(f"{symmetric_difference_set=}")

# Check if set_1 and set_2 have no items in common
print(f"{set_1.isdisjoint({7, 8})=}")

# Comprehensions works with sets:
# Create a set of even squares of numbers from 0 to 9:
squares = {x**2 for x in range(10) if x % 2 == 0}  # Only include even numbers
print(f"{squares=}")

# Unpacking a set:
a, b, c = {1,2,3}
print(f"{a=}, {b=}, {c=}")

# If you want to prevent adding or removing items from a set, you can use a frozenset.
my_frozenset = frozenset([1, 2, 3, 'four', (5, 6)])
print(f"{my_frozenset=}")
```

## Dictionaries

A dictionary is an ordered collection of key-value pairs. They are mutable. You can add, remove, and modify key-value pairs in a dictionary. Dictionaries are defined by using curly braces `{}` and separating key-value pairs with commas. A key-value pair is written as `key: value`. A dictionary can contain items of different types, but the keys must be immutable and unique (like numbers, strings, and tuples). Dictionaries are not indexed, which means you cannot access items by their position. You can access values in a dictionary by using their keys. The syntax is JSON like, so it is perfect to use it if you work with JSON data.

```python
# Example of a dictionary:
my_dict = {'one': 1, 'two': 2, 'three': 3, 'four': 'four', 'five': [5, 6]}
print(f"{my_dict=}")

# Length of the dictionary:
print(f"{len(my_dict)=}")

# You can create dictionaries or cast other data structures to dictionaries with the `dict()` function:
my_tuple = (('one', 1), ('two', 2), ('three', 3))
my_dict_from_tuple = dict(my_tuple)
print(f"{my_dict_from_tuple=}")

# Access values in a dictionary by using their keys:
print(f"{my_dict['one']=}")  # Access value by key 'one'

# Another way with default value if the key is not found:
print(f"{my_dict.get('narf', 'default value')=}")

# Keys, values, and items in a dictionary:
print(f"{my_dict.keys()=}")  # Get all keys in the dictionary
print(f"{my_dict.values()=}")  # Get all values in the dictionary

for key, value in my_dict.items():
    print(f"{key=}, {value=}")

# in checks if a key is in the dictionary:
print(f"{'one' in my_dict=}")

# Change items in the dictionary:
my_dict['one'] = 'one'
print(f"{my_dict=}")

# Insert items in the dictionary:
my_dict['not_existing'] = 'not existing'
print(f"{my_dict=}")

# Remove items from the dictionary by key:
# The `pop()` method removes and returns the item with the specified key. If the key is not found, it returns the default value if provided.
# popitem() will remove the last item.
removed_item = my_dict.pop('not_existing', None)  # Remove the item with key 'not_existing'
print(f"{removed_item=}")
print(f"{my_dict=}")

# An alternative is:
del my_dict["one"]

# Join two dictionaries:
another_dict = {'eight': 8, 'nine': 9}
joined_dict = {**my_dict, **another_dict}  # Merge two dictionaries
print(f"{joined_dict=}")

# Comprehensions is a concise way to create dictionaries:
# Create a dictionary of even squares of numbers from 0 to 9:
squares_dict = {x: x**2 for x in range(10) if x % 2 == 0}  # Only include even numbers
print(f"{squares_dict=}")
```
