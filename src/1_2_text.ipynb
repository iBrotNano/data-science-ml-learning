## Count Words

With the `Counter` class from the `collections` module you can easily count the number of occurrences of each word in a text. This can be useful for various applications, such as analyzing the frequency of words in a document or creating a word cloud.

```python
from collections import Counter

text = """Fluffum schnipsum wackelito banana-dolor sit amet, kicher elit.
Pinguinibus hüpfendibus quantum quatschum, spaghetti tempor blubb.
Kartoffel maximus ploppity pling, senf und salsa con gaudi.
Nudeltronic wummsi deluxe, ipsum klabauter randomicus.
Miau miau, burrito boom, zebrastreifen mit Glitzerhelm.
Taco-loco schnurpsel di doo, pling-plong philosophicus.
Und am Ende sagt der Toaster: „Ich bin bereit für den Weltfrieden und Marmelade."""

parts = text.split() # splits the text on spaces


c = Counter(parts)
print(c.items())
```

## Replace Substrings

There are some methods to replace subtrings in a string.

```python
text = "Hello"
text = text.replace("Hello", "Hi")
print(text)

text = """Die Kräfte knallen, die Fäden verflechten sich, ein Klangteppich voller Funkelfragmente entsteht.
Man könnte fast glauben, ein Kakadu führe ein Fanfarenkonzert auf."""

table = str.maketrans("KFkf", "FKfk")
x = text.translate(table)
print(x)
```

## Regular Expressions

Regular expressions are a powerful tool for searching and manipulating strings based on specific patterns. They can be used for tasks such as validating input, extracting information, or replacing parts of a string. The `re` module in Python provides functions to work with regular expressions.

```python
import re

# Search if a word is in the text:
x = re.search("Narf", text)
print(x) # None

x = re.search("Fluffum", text) # Returns the first match of the pattern in the text

print(type(x),x) # The type is a match object.
print(bool(x)) # You can convert it into a boolean.

# Match objects have some useful methods to get more information about the match:
print(x.span()) # The span method returns the start and end positions of the match in the text. Here it is (0, 7) because "Fluffum" starts at position 0 and ends at position 7 (not included).
print(x.group()) 

# The group method returns the part of the string where there was a match (here trivial, the search string). This will change when the pattern gets more complex.
x = re.search(".ipsum", text)
print(x.group()) # There is one variable character before "ipsum", so it matches "nipsum" in the text.
```

## Pattern Syntax

Regular expressions use a specific syntax to define patterns.

| Character | Description | Example |
|---|---|---|
| `[]` | A set of characters | `"[a-m]"` |
| `\` | Signals a special sequence (can also be used to escape special characters) | `"\d"` |
| `.` | Any character (except newline character) | `"he..o"` |
| `^` | Starts with | `"^hello"` |
| `$` | Ends with | `"planet$"` |
| `*` | Zero or more occurrences | `"he.*o"` |
| `+` | One or more occurrences | `"he.+o"` |
| `?` | Zero or one occurrences | `"he.?o"` |
| `{}` | Exactly the specified number of occurrences | `"he.{2}o"` |
| `\|` | Either or | `"falls\|stays"` |
| `()` | Capture and group |  |

There are some special characters to specify character classes:

| Character | Description | Example |
|---|---|---|
| `\A` | Returns a match if the specified characters are at the beginning of the string | `"\AThe"` |
| `\b` | Returns a match where the specified characters are at the beginning or at the end of a word (the `r` in the beginning makes sure the string is treated as a raw string) | `r"\bain"` / `r"ain\b"` |
| `\B` | Returns a match where the specified characters are present, but NOT at the beginning (or at the end) of a word (the `r` in the beginning makes sure the string is treated as a raw string) | `r"\Bain"` / `r"ain\B"` |
| `\d` | Returns a match where the string contains digits (numbers from 0-9) | `"\d"` |
| `\D` | Returns a match where the string DOES NOT contain digits | `"\D"` |
| `\s` | Returns a match where the string contains a white space character | `"\s"` |
| `\S` | Returns a match where the string DOES NOT contain a white space character | `"\S"` |
| `\w` | Returns a match where the string contains any word characters (characters from a to Z, digits from 0-9, and the underscore `_` character) | `"\w"` |
| `\W` | Returns a match where the string DOES NOT contain any word characters | `"\W"` |
| `\Z` | Returns a match if the specified characters are at the end of the string | `"Spain\Z"` |

You can also define sets of characters:

| Set | Description |
|---|---|
| `[arn]` | Returns a match where one of the specified characters (`a`, `r`, or `n`) is present |
| `[a-n]` | Returns a match for any lower case character, alphabetically between `a` and `n` |
| `[^arn]` | Returns a match for any character EXCEPT `a`, `r`, and `n` |
| `[0123]` | Returns a match where any of the specified digits (`0`, `1`, `2`, or `3`) are present |
| `[0-9]` | Returns a match for any digit between `0` and `9` |
| `[0-5][0-9]` | Returns a match for any two-digit numbers from `00` to `59` |
| `[a-zA-Z]` | Returns a match for any character alphabetically between `a` and `z`, lower case OR upper case |
| `[+]` | In sets, `+`, `*`, `.`, `|`, `()`, `$`, `{}` have no special meaning, so `[+]` means: return a match for any `+` character in the string |

The regex object has some methods to work with regular expressions:

| Function | Description |
|---|---|
| `findall` | Returns a list containing all matches |
| `search` | Returns a Match object if there is a match anywhere in the string |
| `split` | Returns a list where the string has been split at each match |
| `sub` | Replaces one or many matches with a string |
| `match` | Checks for a match only at the beginning of the string. If the whole string matches the pattern |

```python
# Example of a regular expression pattern:
text = """Die Kräfte knallen, die Fäden verflechten sich, ein Klangteppich voller Funkelfragmente entsteht.
Man könnte fast glauben, ein Kakadu führe ein Fanfarenkonzert auf."""

pattern = r"\b\w+en\b" # This pattern matches any word that ends with "um". The \b means a word boundary, \w+ means one or more word characters, and um is the literal string "um". So it will match words like "Fluffum", "schnipsum", "wackelito", etc.
matches = re.findall(pattern, text) # The findall function returns all non-overlapping matches of the pattern in the text as a list of strings.
print(matches)
```
